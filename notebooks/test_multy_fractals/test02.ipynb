```python
import numpy as np
from matplotlib import pyplot as plt
from multifractal_analysis.cascade_simulations import beta_sym, alpha_sym
from multifractal_analysis.fractal_dimension import is_power_of_2
from seaborn import rugplot, barplot
from multifractal_analysis.fractal_dimension import get_fractal_dimension
```

```python
# Run the alpha simulation
N= 16
c = 0.4
gamma_boost = 0.2
beta = alpha_sym(N, c , gamma_boost)

# Find the fractal Dimension
sol = get_fractal_dimension(beta, 0.1)

# Plot both side by side
figure = plt.figure()
figure.set_size_inches((16,8))
ax = figure.add_subplot(1,2,1)
ax.plot(beta)
ax = figure.add_subplot(1,2,2)
ax.plot(sol.ypoints,sol.xpoints,'.')
x = (min(sol.xpoints),max(sol.xpoints))
y = (sol.angular_coef*x[0] + sol.linear_coef, sol.angular_coef*x[1] + sol.linear_coef)
ax.plot(sol.ypoints,sol.xpoints,'.')
ax.plot(y,x)
print(f"Dimension = {sol.angular_coef}")
```
